```python
import pandas as pd
```

```python
fd = pd.read_csv('../data/container_file_descriptors.csv') # time, pod, mean_gauge
fs = pd.read_csv('../data/container_fs_usage_bytes.csv') # time, pod, mean_gauge
sk = pd.read_csv('../data/container_sockets.csv') # time, pod, mean_gauge
th = pd.read_csv('../data/container_threads.csv') # time, pod, mean_gauge
cm = pd.read_csv('../data/cpu_memory_usage.csv') # time, pod_name, mean_cpu_usage_nanocores, mean_memory_working_set_bytes, mean_memory_rss_bytes
nw = pd.read_csv('../data/k8s_pod_network.csv') # time, pod_name, mean_rx_bytes, mean_rx_errors, mean_tx_bytes, mean_tx_errors
```

```python
fd = fd.loc[fd.namespace=='drive']
inner = pd.merge(fd, fs, left_on=['time','pod'], right_on=['time','pod'], how='inner')
inner = inner.loc[:,['time', 'pod', 'mean_gauge_x', 'mean_gauge_y']]
inner.rename(columns={'mean_gauge_x':'fd', 'mean_gauge_y':'fs_usage'}, inplace=True)
inner = inner.dropna()
inner
```

```python
inner = pd.merge(inner, sk, left_on=['time','pod'], right_on=['time','pod'], how='inner')
inner
```

```python
inner = inner.loc[:,['time', 'pod', 'fd', 'fs_usage', 'mean_gauge']]
inner.rename(columns={'mean_gauge':'socket'}, inplace=True)
inner = inner.dropna()
inner
```

```python
inner = pd.merge(inner, th, left_on=['time','pod'], right_on=['time','pod'], how='inner')
inner
```

```python
inner = inner.loc[:,['time', 'pod', 'fd', 'fs_usage', 'socket', 'mean_gauge']]
inner.rename(columns={'mean_gauge':'thread'}, inplace=True)
inner = inner.dropna()
inner
```

```python
inner = pd.merge(inner, cm, left_on=['time','pod'], right_on=['time','pod_name'], how='inner')
inner
```

```python
inner = inner.loc[:,['time', 'pod', 'fd', 'fs_usage', 'socket', 'thread', 'mean_cpu_usage_nanocores', 'mean_memory_working_set_bytes', 'mean_memory_rss_bytes']]
inner.rename(columns={'mean_cpu_usage_nanocores':'cpu', 'mean_memory_working_set_bytes':'memory', 'mean_memory_rss_bytes':'rss'}, inplace=True)
inner = inner.dropna()
inner
```

```python
inner = pd.merge(inner, nw, left_on=['time','pod'], right_on=['time','pod_name'], how='inner')
inner
```

```python
# mean_rx_bytes, mean_rx_errors, mean_tx_bytes, mean_tx_errors
inner = inner.loc[:,['time', 'pod', 'fd', 'fs_usage', 'socket', 'thread', 'cpu', 'memory', 'rss', 'mean_rx_bytes', 'mean_rx_errors', 'mean_tx_bytes', 'mean_tx_errors']]
inner.rename(columns={'mean_rx_bytes':'rx_bytes', 'mean_rx_errors':'rx_errors', 'mean_tx_bytes':'tx_bytes', 'mean_tx_errors':'tx_errors'}, inplace=True)
inner = inner.dropna()
inner
```

```python
inner.index.names = ['id']
inner['rx_errors'] = inner['rx_errors'].astype(float)
inner['tx_errors'] = inner['tx_errors'].astype(float)
inner.head()
```

```python
inner.to_csv('../data/merged_pod_data_v2.csv')
```
